```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Display settings
pd.set_option('display.max_columns', None)
print("Libraries loaded successfully.")
```

```python
# Load the dataset
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print("Dataset loaded successfully.")
print(f"Shape: {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# Preview the first 5 rows
df.head()
```

```python
# Check column names and data types
print("Column Names and Data Types:")
print("=" * 40)
print(df.dtypes)
```

```python
# Check for missing values
print("Missing Values per Column:")
print("=" * 40)
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage': missing_pct})
print(missing_df[missing_df['Missing Count']>0])
print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# TotalCharges is stored as text - convert and expose hidden blanks 
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Now check again
hidden_missing = df['TotalCharges'].isnull().sum()
print(f"Hidden blank values in TotalCharges': {hidden_missing}")
print(f"These represent {round(hidden_missing/len(df)*100, 2)}% of customers")
```

```python
# These 11 customers have a tenure of 0 - they just joined, so TotalCharges = 0
df['TotalCharges'].fillna(0, inplace=True)

print("TotalCharges missing values after fix:", df['TotalCharges'].isnull().sum())
print("Data type now:", df['TotalCharges'].dtype)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

```python
# Statistical summary of numerical columns
print("Summary Statistics:")
df.describe()
```

```python
# Check churn distribution
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100

print("Churn Distribution:")
print("=" * 30)
print(f"No  (Active):  {churn_counts['No']}  ({churn_pct['No']:.1f}%)")
print(f"Yes (Churned): {churn_counts['Yes']}  ({churn_pct['Yes']:.1f}%)")
```

```python
# Summary of data understanding findings
print("=" * 45)
print("DATA UNDERSTANDING SUMMARY")
print("=" * 45)
print(f"Total Customers:        {len(df):,}")
print(f"Total Features:         {df.shape[1]}")
print(f"Duplicate Rows:         0")
print(f"Missing Values (fixed): 11 (TotalCharges)")
print(f"Active Customers:       5,174 (73.5%)")
print(f"Churned Customers:      1,869 (26.5%)")
print(f"Avg Monthly Charges:    ${df['MonthlyCharges'].mean():.2f}")
print(f"Avg Tenure:             {df['tenure'].mean():.0f} months")
print(f"Max Tenure:             {df['tenure'].max()} months")
print("=" * 45)

# Save cleaned dataset for use in next notebooks
df.to_csv('../data/telecomx_clean.csv', index=False)
print("\nClean dataset saved to /data/telecomx_clean.csv")
```
